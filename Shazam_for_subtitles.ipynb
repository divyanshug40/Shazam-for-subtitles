```python
import zipfile
import io

count = 0

def decode_method(binary_data):
    global count
    # Decompress the binary data using the zipfile module
    # print(count, end=" ")
    count += 1
    with io.BytesIO(binary_data) as f:
        with zipfile.ZipFile(f, 'r') as zip_file:
            # Assuming there's only one file in the ZIP archive
            subtitle_content = zip_file.read(zip_file.namelist()[0])

    # Now 'subtitle_content' should contain the extracted subtitle content
    return subtitle_content.decode('latin-1')  # Assuming the content is UTF-8 encoded text
```

```python
df['file_content'] = df['content'].apply(decode_method)
df.drop(columns=['content'], inplace=True)
```

```python
import pandas as pd
import re


# Function to clean subtitle data
def clean_subtitle_text(text):


    # Remove timestamps (any text that looks like '00:00:06,000 --> 00:00:12,074')
    cleaned_text = re.sub(r'\d{2}:\d{2}:\d{2},\d{3} --> \d{2}:\d{2}:\d{2},\d{3}', '', text)

    # Remove lines that contain numbers (like the subtitle index '1', '2', etc.)
    cleaned_text = re.sub(r'^\d+\s*$', '', cleaned_text, flags=re.MULTILINE)

    # Remove extra whitespaces and empty lines
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()

    # Remove <i> and </i> tags
    cleaned_text = re.sub(r'</?i>', '', cleaned_text)

    return cleaned_text

# Apply the cleaning function to the 'file_content' column



```

```python
df['cleaned_content'] = df['file_content'].apply(clean_subtitle_text)
```

# Extracting Subtitles given a movie name

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/drive/MyDrive/Innomatics Task 8/eng_subtitles_database.db')
cursor = conn.cursor()

# Fetch table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
print(cursor.fetchall())

# Get the column information for the 'zipfiles' table
cursor.execute("PRAGMA table_info('zipfiles')")
cols = cursor.fetchall()

# Modify the SQL query to pull out only the row for the specific movie
query = f"""SELECT * FROM zipfiles WHERE name = ? LIMIT 1"""

# Use parameterized query to avoid SQL injection
df = pd.read_sql_query(query, conn, params=(movie_name,))

# Print the result to check
print(df)

# Close the connection
conn.close()
```

```python
import zipfile
import io

count = 0

def decode_method(binary_data):
    global count
    # Decompress the binary data using the zipfile module
    # print(count, end=" ")
    count += 1
    with io.BytesIO(binary_data) as f:
        with zipfile.ZipFile(f, 'r') as zip_file:
            # Assuming there's only one file in the ZIP archive
            subtitle_content = zip_file.read(zip_file.namelist()[0])

    # Now 'subtitle_content' should contain the extracted subtitle content
    return subtitle_content.decode('latin-1')  # Assuming the content is UTF-8 encoded text
```

```python
df['file_content'] = df['content'].apply(decode_method)

df.head()
```

```python
import re

file_name = movie_name

# Regex pattern to handle series (season, episode, episode name)
series_pattern = r'\.s(\d{2})\.e(\d{2})\.([a-zA-Z0-9]+)'

# Extracting info for the second title (series)
match = re.search(series_pattern, file_name)

is_series = False

if match:
    is_series = True
    # Extract season, episode, and episode name
    season = match.group(1)  # Season
    episode = match.group(2)  # Episode
    episode_name = match.group(3)  # Episode name

    # Remove the season, episode, and episode name, replace dots with spaces
    file_name = file_name.replace(match.group(0), '')

# Regex pattern for movies (captures the name before the year)
movie_pattern = r'([a-zA-Z0-9.]+(?:\.[a-zA-Z0-9]+)*)\.\((\d{4})\)\.'

# Extracting info for the first title (movie)
match_movie = re.search(movie_pattern, file_name)

if match_movie:
    movie_name = match_movie.group(1).replace('.', ' ').title()
    movie_year = match_movie.group(2)
else:
    movie_name = "Unknown"
    movie_year = "Unknown"

    # Print movie details with dots replaced by spaces
print(f"Movie Name: {movie_name}")
if is_series:
    print(f"Season: {season}, Episode: {episode}")
    print(f"Episode Name: {episode_name.replace('.', ' ').title()}")
print(f"Year: {movie_year}")
```

```python
# Create SRT file

import re

# Assuming your content is in a variable (e.g., df['file_content'][0])
content = df['file_content'][0]  # Replace this with your actual content variable

# Clean up the initial BOM (Byte Order Mark) if it exists
content = content.replace('ï»¿', '')

# Split the content by subtitle blocks
subtitle_blocks = content.strip().split('\n\n')

# Open a file to write the subtitles
with open('output.srt', 'w') as f:
    for i, block in enumerate(subtitle_blocks):
        # Split each block into individual parts (index, time, text)
        lines = block.split('\n')
        index = lines[0]
        time_range = lines[1]
        text = '\n'.join(lines[2:])

        # Write the index, time range, and text in the SRT format
        f.write(f"{index}\n")
        f.write(f"{time_range}\n")
        f.write(f"{text}\n\n")

print("SRT file created successfully.")
```
